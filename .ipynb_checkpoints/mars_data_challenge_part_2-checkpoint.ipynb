```python
# Import dependencies 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager

# Set up Browser with Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Have the browser navigate to the website and copy the content
url = 'https://data-class-mars-challenge.s3.amazonaws.com/Mars/index.html'
browser.visit(url)
html = browser.html
soup = soup(html, 'html.parser')
```

```python
# Find the table
table = soup.find('table', class_='table')
table
```

```python
# Find the header
table_header = table.find_all('th')

headers=[]
for header_ele in table_header:
    title=header_ele.text
    headers.append(title)
    
headers
```

```python
mars_df=pd.DataFrame(columns=headers)
mars_df
```

```python
# Find the row data 
for row in table.find_all('tr',class_='data-row'):
        data = row.find_all('td')
        row_data=[td.text.strip() for td in data]
        lenghth=len(mars_df)
        mars_df.loc[lenghth]=row_data
        
mars_df
```

```python
# Close the session 
browser.quit()
```

```python
# Check data types
mars_df.dtypes
```

```python
# Change data types 
mars_df['id']=mars_df['id'].astype(int)
mars_df['terrestrial_date']=pd.to_datetime(mars_df['terrestrial_date'])
mars_df['sol']=mars_df['sol'].astype(int)
mars_df['ls']=mars_df['ls'].astype(int)
mars_df['month']=mars_df['month'].astype(int)
mars_df['min_temp']=mars_df['min_temp'].astype(float)
mars_df['pressure']=mars_df['pressure'].astype(float)

# Check data types again 
mars_df.info()
```

```python
#How many months exist on Mars?
number_of_months = mars_df['month'].nunique()
number_of_months
```

```python
#How many Martian (and not Earth) days worth of data exist in the scraped dataset?

days_data =mars_df['sol'].nunique()
days_data
```

```python
#What are the coldest and the warmest months on Mars (at the location of Curiosity)? 
mars_df_by_month=mars_df.groupby('month')
mars_df_month_temperature=mars_df_by_month.mean().loc[:,['min_temp']]
mars_df_month_temperature
```

```python
# Plot avg_pressure
x=range(1,13)
plt.bar(x,mars_df_month_temperature['min_temp'])

# Create labels for the x and y axes.
plt.xlabel("Months")
plt.ylabel("min_temp")

# Create title 
plt.title('Avergae temperature by Month')
```

```python
min_temperature = mars_df_month_temperature['min_temp'].min()
month_min_temperature=mars_df_month_temperature.loc[mars_df_month_temperature['min_temp']==min_temperature]
month_min_temperature
```

```python
max_temperature = mars_df_month_temperature['min_temp'].max()
month_max_temperature=mars_df_month_temperature.loc[mars_df_month_temperature['min_temp']==max_temperature]
month_max_temperature
```

```python
#Which months have the lowest and the highest atmospheric pressure on Mars? 
mars_df_by_month=mars_df.groupby('month')
mars_df_month_pressure=mars_df_by_month.mean().loc[:,['pressure']]
mars_df_month_pressure
```

```python
# Plot avg_pressure
x=range(1,13)
plt.bar(x,mars_df_month_pressure['pressure'])

# Create labels for the x and y axes.
plt.xlabel("Months")
plt.ylabel("pressure")

# Create title 
plt.title('Avergae pressure by Month')
```

```python
min_pressure = mars_df_month_pressure['pressure'].min()
month_min_pressure=mars_df_month_pressure.loc[mars_df_month_pressure['pressure']==min_pressure]
month_min_pressure
```

```python
max_pressure = mars_df_month_pressure['pressure'].max()
month_max_pressure=mars_df_month_pressure.loc[mars_df_month_pressure['pressure']==max_pressure]
month_max_pressure
```

```python
#how many terrestrial (Earth) days exist in a Martian year?
#Finding the original Sun longitude
original_longitude=mars_df['ls'].loc[0]
original_longitude
```

```python
#Sun longitude will be in the same location after approx 1 year. Need to find the rows where longitude value is equal to original longitude
same_longitude = mars_df.loc[mars_df['ls']==original_longitude]
same_longitude
```

```python
import datetime as dt
from datetime import datetime

x=same_longitude['terrestrial_date'].loc[0]
date_after_one_mars_year = same_longitude['terrestrial_date'].iloc[1]
one_mars_year= date_after_one_mars_year-x
one_mars_year
```
