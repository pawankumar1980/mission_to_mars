```python
# Imports
import json
import pandas as pd


from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Browser with Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit the Mars NASA news site
url = 'https://redplanetscience.com'
browser.visit(url)
# Optional delay for loading the page
browser.is_element_present_by_css('div.list_text', wait_time=1)
```

```python
html = browser.html
news_soup = soup(html, 'html.parser')
news_soup
```

```python
#Scrapping the title text for each article
news_titles=news_soup.find_all('div',class_='content_title')

x=1
for title in news_titles:
    print(x,".",title.text)
    x=x+1
```

```python
#Scrapping the Summary text for each article
news_summaries =news_soup.find_all('div', class_ ='article_teaser_body')

y=1
for summary in news_summaries:
    print(y,".",summary.text)
    y=y+1
    
```

```python
# Slide_elem is the parent element that holds all the elements within it
slide_elem = news_soup.find_all('div',class_='list_text')

# Itereate through all the element to scrap the titles and the summary for each article
mars_list = []
for elem in slide_elem:
    mars_data={"title":"","preview":""}
    title_elem=elem.find('div',class_='content_title')
    title = title_elem.get_text()
    summary_elem= elem.find('div', class_='article_teaser_body')
    summary=summary_elem.get_text()
    mars_data["title"] = title
    mars_data["preview"] = summary
    mars_list.append(mars_data)
print(mars_list)
```

```python
# Export Python list to JSON file
json_mars = json.dumps(mars_list)
print(json_mars)
```

```python
# Save json as csv file
df=pd.read_json(json_mars)
df.to_csv('mars.csv',index=False)
```

```python
# Quit browser 
browser.quit()
```
